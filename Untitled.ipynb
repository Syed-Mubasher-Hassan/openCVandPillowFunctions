# image manipulations with Pillow

```python
img=Image.open(r"F:\css pic\p (20).jpg")
img_f=ImageOps.flip(img)
img_m=ImageOps.mirror(img)

fig=plt.figure(figsize=(14,12))
# setting values to rows and column variables
# rows = 1
# columns = 3

# Adds a subplot at the 1st position
fig.add_subplot(1, 3, 1)
plt.imshow(img)
plt.title('image')
# Adds a subplot at the 1st position
fig.add_subplot(1, 3, 2)
plt.imshow(img_f)
plt.title('image flip')
# Adds a subplot at the 1st position
fig.add_subplot(1, 3, 3)
plt.imshow(img_m)
plt.title('image mirror')
plt.axis('off')
```

# image munipulation with Cv2

```python
img=cv2.imread(r"F:\css pic\p (20).jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_f=cv2.flip(img,0)
img_r=cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)

fig=plt.figure(figsize=(14,12))
rows=1
columns=3

fig.add_subplot(rows,columns,1)
plt.imshow(img)
plt.title('image')

fig.add_subplot(rows,columns,2)
plt.imshow(img_f)
plt.title('image flip')

fig.add_subplot(rows,columns,3)
plt.imshow(img_r)
plt.title('image rotate')
```
